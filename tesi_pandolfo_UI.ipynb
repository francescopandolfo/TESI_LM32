```python
import numpy as np
import pandas as pd
import requests
import matplotlib.pyplot as plt

url_get_ts = "http://10.200.100.235:8001/gettimeseries/query"
query_body = {"stationcode" : "EIIV",
              "startdate" : "2024-01-01",
              "enddate" : "2024-01-31", 
              "last" : "2d" #commentare se non si usa. #non andare oltre 1h, oppure esprimere in "d" (day)
             } 
headers =  {"Content-Type":"application/json"}
response = requests.post(url_get_ts, json=query_body, headers=headers)
data = response.json()
df = pd.DataFrame(data['dataframe'])
df
```
