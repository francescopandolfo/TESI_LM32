```python
df = _df1
df1 = _df1[_df1['ppm'].isna()==False]
df2 = _df2[_df2['ppm'].isna()==False]
df3 = _df3[_df3['ppm'].isna()==False]
df4 = _df4[_df4['ppm'].isna()==False]
df5 = _df5[_df5['ppm'].isna()==False]
df6 = _df6[_df6['ppm'].isna()==False]
df7 = _df7[_df7['ppm'].isna()==False]
df8 = _df8[_df8['ppm'].isna()==False]
df9 = _df9[_df9['ppm'].isna()==False]
df10 = _df10[_df10['ppm'].isna()==False]
df11 = _df11[_df11['ppm'].isna()==False]
df12 = _df12[_df12['ppm'].isna()==False]
df13 = _df13[_df13['ppm'].isna()==False]
df14 = _df14[_df14['ppm'].isna()==False]
df15 = _df15[_df15['ppm'].isna()==False]
df16 = _df16[_df16['ppm'].isna()==False]
df17 = _df17[_df17['ppm'].isna()==False]
df18 = _df18[_df18['ppm'].isna()==False]
df19 = _df19[_df19['ppm'].isna()==False]
df20 = _df20[_df20['ppm'].isna()==False]
```

```python
df['edam_ecne_eplu'] = df1['ppm'].str.replace(',','.').astype(float)
df['edam_ecor_emcn'] = df2['ppm'].str.replace(',','.').astype(float)
df['edam_emsg_eplu'] = df3['ppm'].str.replace(',','.').astype(float)
df['edam_epdn_ecne'] = df4['ppm'].str.replace(',','.').astype(float)
df['edam_epdn_emcn'] = df5['ppm'].str.replace(',','.').astype(float)
df['eint_ecpn_emgl'] = df6['ppm'].str.replace(',','.').astype(float)
df['eint_esln_emgl'] = df7['ppm'].str.replace(',','.').astype(float)
df['eint_esln_espc'] = df8['ppm'].str.replace(',','.').astype(float)
df['emeg_ecpn_emgl'] = df9['ppm'].str.replace(',','.').astype(float)
df['emfn_ecor_ecri'] = df10['ppm'].str.replace(',','.').astype(float)
df['emfn_ecor_emcn'] = df11['ppm'].str.replace(',','.').astype(float)
df['emfn_emcn_epdn'] = df12['ppm'].str.replace(',','.').astype(float)
df['epdn_ecne_ecpn'] = df13['ppm'].str.replace(',','.').astype(float)
df['epdn_eint_ecpn'] = df14['ppm'].str.replace(',','.').astype(float)
df['epdn_eint_espc'] = df15['ppm'].str.replace(',','.').astype(float)
df['epdn_emfn_espc'] = df16['ppm'].str.replace(',','.').astype(float)
df['eplu_ecne_ecpn'] = df17['ppm'].str.replace(',','.').astype(float)
df['eplu_emeg_ecpn'] = df18['ppm'].str.replace(',','.').astype(float)
df['eplu_emeg_emal'] = df19['ppm'].str.replace(',','.').astype(float)
df['eplu_emsg_emal'] = df20['ppm'].str.replace(',','.').astype(float)

format = '%d/%m/%Y'
df['date'] = pd.to_datetime(df['day'], format=format)
df = df.set_index('date')

df = df.drop(columns=['ppm'])
df = df.drop(columns=['day'])
#df['day'].astype('time').dtypes
print(df.info())
```
